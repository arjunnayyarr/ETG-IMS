```python
#Add a new product to the inventory. 
import json
fd = open('records.json')
r = fd.read()
records = json.loads(r)

num_new_products = int(input("How many products do you want to add to the inventory? "))
for i in range(num_new_products):
    print("Enter details about product " +str(num_new_products))
    new_prod_id = str(input("Enter new product id: "))
    new_prod_name = str(input("Enter new product name: "))
    new_prod_price = int(input("Enter new product price: "))
    new_prod_quantity = int(input("Enter new product quantity: "))
    print("")
    print("************************************************************")
    print("")

    records[new_prod_id] = {"name" : new_prod_name, "price" : new_prod_price, "quantity" : new_prod_quantity}
    
    
    
product_js = json.dumps(records)
fd = open('records.json','w')
fd.write(product_js)
fd.close()

print("INVENTORY: ")
print("************************************************************")
records
```
